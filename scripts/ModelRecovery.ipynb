## Model Recovery
Rerun model using HDDM regression - but with the least number of parameters


```python
import pandas as pd
import matplotlib.pyplot as plt
import hddm
import numpy as np
import patsy 

samples = 5000

data_type = 'SimpleFull' # Replace with SimpleDrift or SimpleBias to fit data generated by those models

model_DIC = pd.DataFrame(np.nan, index=range(1,101), columns=['SimpleFull','SimpleBias','SimpleDrift'])
```

```python
for sim in range(1,101):

    print 
    print 'Simulation', sim
    
    data = hddm.load_csv('../data/model_outputs/simulations/model_recovery/' + 
                         data_type + '/Sim' + str(sim) + '.csv')
    
    def z_link_func(x, data=data):
        return 1 / (1 + np.exp(-(x)))
    
    ###### Full Model #####
    model_type = 'SimpleFull'
    print "Model =", model_type
    v_reg = {'model': 'v ~ 1 + condition + stim', 'link_func': lambda x: x}
    z_reg = {'model': 'z ~ 1 + condition', 'link_func': z_link_func}

    reg_descr = [v_reg, z_reg]

    thisModel = hddm.models.HDDMRegressor(data, reg_descr,
                                           bias=True, p_outlier=0.05,
                                           group_only_regressors=False)

    thisModel.sample(samples, burn=samples/10, thin=2) 

    # save data
    model_DIC.loc[sim]['SimpleFull'] = thisModel.dic
    
    # Bias Model
    model_type = 'SimpleBias'
    print
    print "Model =", model_type
    v_reg = {'model': 'v ~ 1 + stim', 'link_func': lambda x: x}
    z_reg = {'model': 'z ~ 1 + condition', 'link_func': z_link_func}

    reg_descr = [v_reg, z_reg]

    thisModel = hddm.models.HDDMRegressor(data, reg_descr,
                                           bias=True, p_outlier=0.05,
                                           group_only_regressors=False)

    thisModel.sample(samples, burn=samples/10, thin=2) 
    
    # save data
    model_DIC.loc[sim]['SimpleBias'] = thisModel.dic

    # Drift Model
    model_type = 'SimpleDrift'
    print
    print "Model =", model_type
    v_reg = {'model': 'v ~ 1 + condition + stim', 'link_func': lambda x: x}
    z_reg = {'model': 'z ~ 1', 'link_func': z_link_func}

    reg_descr = [v_reg, z_reg]

    thisModel = hddm.models.HDDMRegressor(data, reg_descr,
                                           bias=True, p_outlier=0.05,
                                           group_only_regressors=False)

    thisModel.sample(samples, burn=samples/10, thin=2) 
    
    # save data
    model_DIC.loc[sim]['SimpleDrift'] = thisModel.dic
    
    model_DIC.to_csv('../data/model_outputs/simulations/model_recovery/' + 
                         data_type + '.csv')
```
